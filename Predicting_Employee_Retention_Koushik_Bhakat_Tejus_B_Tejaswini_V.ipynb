7.3.1 Plot ROC curve <font color="red">[3 Marks]</font>

```python
# Define ROC function
def plot_roc_curve(y_true, y_pred_prob, title='ROC Curve'):
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
    auc = roc_auc_score(y_true, y_pred_prob)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Diagonal line
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate (Recall)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print(f"AUC Score: {auc:.2f}")
```

7.3.2 Predict on training set at various probability cutoffs <font color="red">[1 Mark]</font>

```python
# Predict on training data by creating columns with different probability cutoffs to explore the impact of cutoff on model performance
# Define a list of cutoff values to evaluate
cutoff_values = [round(x * 0.1, 2) for x in range(1, 10)]  # 0.1 to 0.9

# Initialize a list to store results
cutoff_analysis = []

# Loop through different cutoff values
for cutoff in cutoff_values:
    predicted = train_results['Predicted_Probability'].apply(lambda x: 1 if x > cutoff else 0)
    tn, fp, fn, tp = metrics.confusion_matrix(train_results['Actual_Attrition'], predicted).ravel()
    
    # Calculate sensitivity and specificity
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
    
    cutoff_analysis.append([cutoff, sensitivity, specificity])

# Convert to DataFrame
cutoff_df = pd.DataFrame(cutoff_analysis, columns=['Cutoff', 'Sensitivity', 'Specificity'])

# Display the cutoff analysis
print(cutoff_df)
```

7.3.3 Plot for accuracy, sensitivity, specificity at different probability cutoffs <font color="red">[2 Marks]</font>

```python
# Create a DataFrame to see the values of accuracy, sensitivity, and specificity at different values of probability cutoffs
# Define a list of cutoff values to evaluate
cutoff_values = [round(x * 0.05, 2) for x in range(1, 20)]  # 0.05 to 0.95

# Initialize a list to store results
cutoff_analysis = []

# Loop through different cutoff values
for cutoff in cutoff_values:
    predicted = train_results['Predicted_Probability'].apply(lambda x: 1 if x > cutoff else 0)
    
    # Confusion matrix
    tn, fp, fn, tp = metrics.confusion_matrix(train_results['Actual_Attrition'], predicted).ravel()
    
    # Calculate metrics
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
    
    # Append to list
    cutoff_analysis.append([cutoff, accuracy, sensitivity, specificity])

# Convert to DataFrame
cutoff_df = pd.DataFrame(cutoff_analysis, columns=['Cutoff', 'Accuracy', 'Sensitivity', 'Specificity'])

# Display the DataFrame
print(cutoff_df)
```
